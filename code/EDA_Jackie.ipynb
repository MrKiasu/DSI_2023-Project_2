```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn import preprocessing
from sklearn import model_selection
from sklearn import metrics
from sklearn import linear_model
```

```python
pd.set_option("display.max_rows", 50)
pd.set_option("display.max_columns", 60)
```

```python
df = pd.read_csv("../data/df_clean.csv")
town_region = pd.read_csv("../data/town_region.csv")
```

```python
df = pd.merge(df, town_region, on = "town", how = "left")
```

```python
df.drop("Unnamed: 0", axis = 1, inplace = True)
```

```python
numeric_columns = ["tranc_year", "floor_area_sqm", "hdb_age", "relative_height", "percent_sold", "mall_nearest_distance", "hawker_nearest_distance", "mrt_nearest_distance", "pri_sch_nearest_distance"]
categorical_columns = ["flat_type", "town"]
df_cat = pd.get_dummies(df[categorical_columns], drop_first = True, dtype = "int")
df_features = pd.merge(df[numeric_columns], df_cat, how = "inner", left_index = True, right_index = True)
```

```python
len(numeric_columns)
```

```python
df_corr = pd.merge(df_features, df["resale_price"], how = "inner", left_index = True, right_index = True)
```

```python
df_corr.corr()
```

```python
y = df['resale_price']
x_train, x_test, y_train, y_test = model_selection.train_test_split(df_features, y, random_state=42)
lr = linear_model.LinearRegression()
lr.fit(x_train,y_train)
y_pred = lr.predict(x_train)
print(f"Train score is {lr.score(x_train,y_train)}")
train_mse = metrics.mean_squared_error(y_train, y_pred)
print(f"Train RMSE is {(train_mse)**0.5}")

lr = linear_model.LinearRegression()
lr.fit(x_test,y_test)
y_pred = lr.predict(x_test)
print(f"Test score is {lr.score(x_test,y_test)}")
test_mse = metrics.mean_squared_error(y_test, y_pred)
print(f"Test RMSE is {(test_mse)**0.5}")
```
